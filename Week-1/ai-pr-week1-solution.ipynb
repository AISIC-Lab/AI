### Challenge 1 — Student Grade Analyzer
Ask the user for:

- Their name

2. Calculate the average grade.

### Challenge 2 — Shopping Cart Summary
Ask the user to input:

- Three product names
- Their respective prices (`float()`)

Then:

1. Store the data in a dictionary (product → price).
2. Use a `for` loop to calculate the total cost.
3. Apply a 10% discount if the total is greater than 50.
4. Print the cart contents, total cost, and final price after any discount.

````python
cart = {}

for i in range(3):
    product = input("Product name: ")
    price = float(input("Price: "))
    cart[product] = price

total = sum(cart.values())

if total > 50:
    final_price = total * 0.9
else:
    final_price = total

print("\nCart:")
for product, price in cart.items():
    print(product, "->", price)

print(f"Total: {total:.2f}")
print(f"Final price: {final_price:.2f}")
````

### Challenge 3 — Number Statistics Tool
Ask the user for five numbers (`float()`):

1. Store them in a list.
2. Count how many numbers are:

   - Greater than the average

````python
numbers = []

for i in range(5):
    numbers.append(float(input(f"Number {i+1}: ")))

average = sum(numbers) / len(numbers)

positive = 0
negative = 0
above_avg = 0

for n in numbers:
    if n > 0:
        positive += 1
    elif n < 0:
        negative += 1

    if n > average:
        above_avg += 1

print(f"Average: {average:.2f}")
print("Positive numbers:", positive)
print("Negative numbers:", negative)
print("Above average:", above_avg)
````

### Challenge 5 — Library Management Mini-System
Create a simple library management program.

The system should allow:

1. **Book registration and removal**
   - Store books in a dictionary or list of dictionaries (title, author, availability).
   - Allow adding new books and removing existing ones.

2. **Borrowing books**
   - Ask which book the user wants.
   - If available, mark it as borrowed.
   - Otherwise, inform that it is unavailable.

3. **Returning books**
   - Allow marking a borrowed book as returned.

4. Use:
   - Lists and dictionaries to store data
   - `for` loops to search books
   - `if` statements for decision logic
   - Proper input handling (`str`, `int`, `float` if needed)

5. Print clear status messages for each operation.

````python
library = []

while True:
    print("\n1. Add book")
    print("2. Remove book")
    print("3. Borrow book")
    print("4. Return book")
    print("5. Show library")
    print("0. Exit")

    choice = input("Option: ")

    if choice == "1":
        title = input("Title: ")
        author = input("Author: ")
        library.append({"title": title, "author": author, "available": True})

    elif choice == "2":
        title = input("Book to remove: ")
        library = [b for b in library if b["title"] != title]

    elif choice == "3":
        title = input("Book to borrow: ")
        for book in library:
            if book["title"] == title:
                if book["available"]:
                    book["available"] = False
                    print("Borrowed successfully.")
                else:
                    print("Book unavailable.")

    elif choice == "4":
        title = input("Book to return: ")
        for book in library:
            if book["title"] == title:
                book["available"] = True
                print("Returned successfully.")

    elif choice == "5":
        for book in library:
            status = "Available" if book["available"] else "Borrowed"
            print(book["title"], "-", book["author"], "-", status)

    elif choice == "0":
        break
````

### Exercise 2 — Challenge with Functions

Create a small grade management program using functions.
The program should:

1. Create a function get_grades() that:
    - Asks the user for five grades (float()).
    - Stores them in a list.
    - Returns the list.

2. Create another function analyze_grades(grades) that:
    - Calculates the average grade.
    - Determines if the student passed (average ≥ 10).
    - Returns both the average and the result (True/False).

3. In the main program:
    - Ask the student’s name.
    - Call both functions.
    - Print a formatted summary with the name, grades, average, and pass/fail status.

````python
def get_grades():
    grades = []
    for i in range(5):
        grades.append(float(input(f"Grade {i+1}: ")))
    return grades


def analyze_grades(grades):
    average = sum(grades) / len(grades)
    passed = average >= 10
    return average, passed


name = input("Student name: ")

grades = get_grades()
average, passed = analyze_grades(grades)

print(f"\nStudent: {name}")
print("Grades:", grades)
print(f"Average: {average:.2f}")
print("Status:", "Passed" if passed else "Failed")
````

````python
def add_book(library):
    title = input("Title: ")
    author = input("Author: ")
    library.append({"title": title, "author": author, "available": True})
    print("Book added successfully.")


def remove_book(library):
    title = input("Book to remove: ")
    for book in library:
        if book["title"] == title:
            library.remove(book)
            print("Book removed.")
            return
    print("Book not found.")


def borrow_book(library):
    title = input("Book to borrow: ")
    for book in library:
        if book["title"] == title:
            if book["available"]:
                book["available"] = False
                print("Borrowed successfully.")
            else:
                print("Book unavailable.")
            return
    print("Book not found.")


def return_book(library):
    title = input("Book to return: ")
    for book in library:
        if book["title"] == title:
            book["available"] = True
            print("Returned successfully.")
            return
    print("Book not found.")


def show_library(library):
    if not library:
        print("Library is empty.")
        return

    for book in library:
        status = "Available" if book["available"] else "Borrowed"
        print(f"{book['title']} - {book['author']} - {status}")


library = []

while True:
    print("\n1. Add book")
    print("2. Remove book")
    print("3. Borrow book")
    print("4. Return book")
    print("5. Show library")
    print("0. Exit")

    choice = input("Option: ")

    if choice == "1":
        add_book(library)
    elif choice == "2":
        remove_book(library)
    elif choice == "3":
        borrow_book(library)
    elif choice == "4":
        return_book(library)
    elif choice == "5":
        show_library(library)
    elif choice == "0":
        break
````
